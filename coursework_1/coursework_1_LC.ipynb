```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

## 1. Load data and print the first few lines using pandas (10 points).

```python
data = pd.read_csv('BostonHousing.csv')

print(data.loc[0:10,:])
```

## Dataset description
Each row is a case of the housing price. There are 506 cases in total. Each column is an attribute, there are 14 attributes, including:

**crim**: per capita crime rate by town

**zn**: proportion of residential land zoned for lots over 25,000 sq.ft.

**indus**: proportion of non-retail business acres per town

**chas**: Charles River dummy variable (1 if tract bounds river; 0 otherwise)

**nox**: nitric oxides concentration (parts per 10 million)

**rm**: average number of rooms per dwelling

**age**: proportion of owner-occupied units built prior to 1940

**dis**: weighted distances to five Boston employment centres

**rad**: index of accessibility to radial highways

**tax**: full-value property-tax rate per \$10,000

**ptratio**: pupil-teacher ratio by town

**b**: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town

**lstat**: lower status of the population

**medv**: Median value of owner-occupied homes in $1000's

## 2. Simple statistics (10 points).
Print the basic statistics (mean and standard deviation) for the crime rate, nitric oxides concentration, pupil-teacher ratio and median value of owner-occupied homes.

```python
columns = ['crim','nox','ptratio','medv']

avg_vals = data.mean()
std_vals = data.std()
print("Printing average values for the crime rate, nitric oxides concentration\n",
      "pupil-teacher ratio and median value of owner-occupied homes:")
print(avg_vals.loc[columns])
print("Printing standard deviation values for the crime rate, nitric oxides concentration\n",
      "pupil-teacher ratio and median value of owner-occupied homes:")
print(std_vals.loc[columns])
```

## 3. Data visualisation (30 points).
### 3.1 Plot the histogram distribution for each data column (10 points).

```python
fig, axs = plt.subplots(figsize=(12, 12), ncols=4,nrows = 4)
axs = axs.ravel()
for idx,col in enumerate(data.columns):
    axs[idx].hist(data.loc[:,col], color = "red",edgecolor = "black")
    axs[idx].set_title(col)
    axs[idx].set_xlabel("Value")
    axs[idx].set_ylabel("Count")
plt.tight_layout()

axs[14].set_visible(False)
axs[15].set_visible(False)
```

### 3.2 Plot the correlation matrix between the data columns (10 points).

```python
plt.matshow(data.corr())
plt.xticks(range(data.shape[1]), data.columns, rotation=90)
plt.yticks(range(data.shape[1]), data.columns)
plt.show()
```

### 3.3 Plot the house price (the last data column) against each feature (each of the first 13 data columns) (10 points).

```python
fig, axs = plt.subplots(figsize=(12, 12), ncols=4,nrows = 4)
housePrice = data.loc[:, data.columns == 'medv']
rest = data.loc[:, data.columns != 'medv']
axs = axs.ravel()
for idx,col in enumerate(rest.columns):
    axs[idx].scatter(rest.loc[:,col],housePrice, edgecolor = 'black', color = 'green')
    axs[idx].set_title("Relationship between \nhouse price and "+col)
    axs[idx].set_xlabel(col)
    axs[idx].set_ylabel("House price")
plt.tight_layout()

axs[13].set_visible(False)
axs[14].set_visible(False)
axs[15].set_visible(False)

```

## 4. Linear regression (30 points).
### 4.1. Regress the house price against all the features (15 points).

* First, split the whole dataset into a training set and a test set using a pre-defined ratio (80:20 in this case).

* Then, train the linear regression model using the training set.

* Finally, plot the predicted house price on the training set.

The dataset split is provided for consistent evaluation. Please do not change the random_state seed.

```python
X = data.iloc[:, :13]
y = data.iloc[:, 13]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

### 4.2 Quantitatively evaluate the linear model using the root of mean squared error (RMSE) and the R squared (R2) on both the training set and test set (15 points).

```python
#RMSE defined as square root of mean of square differences between prediction and targets
predTrain = reg.predict(X_train)
predTest = reg.predict(X_test)
RMSETrain = np.sqrt(mean_squared_error(y_train,predTrain))
RMSETest = np.sqrt(mean_squared_error(y_test,predTest))
R2Train = r2_score(y_train,predTrain)
R2Test = r2_score(y_test,predTest)

# This question: "Finally, plot the predicted house price on the training set."
# is a bit confusing therefore I am plotting the following things:
# Predicted house price(y-axis) vs. Actual house price for training set(x-axis)

plt.plot(np.linspace(0,50),np.linspace(0,50),color='black',linestyle = 'dashed')
plt.scatter(y_train,predTrain, color = "red", marker = "+")
plt.xlabel("True value")
plt.ylabel("Predicted value")
plt.title("Predicted vs. true value for training set")
plt.show()

print("RMSE_train = "+ str(RMSETrain))
print("RMSE_test = "+ str(RMSETest))
print("\n")
print("R2_train = "+ str(R2Train))
print("R2_test = "+ str(R2Test))
```

## 5. Challenge yourself (20 points)

Previously, we use 13 features to predict the house price. Perhaps some of the features are more relevant to the price, whereas some are less.

### 5.1 Explore the features and develop a linear model for house price prediction using only 5 features as input (10 points).

Hint: either using feature selection or dimensionality reduction.

```python
#store StandardScaler function into variable Scaler
scaler = StandardScaler()

#make scaling "model" for x_train data
scaler.fit(X_train)
#scale variables x_train and x_test. Note: we are scaling x_test with our x_train scaling "model"
# as we do not want to "show" x_test to our model at any time
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)


pca = PCA(n_components=5, svd_solver='full')
#perform PCA on scaled X_train giving your 5 new PC dimensions
pca.fit(X_train)
#print explained variance ratio for these components
print("PCA top 5 components explain", sum(pca.explained_variance_ratio_)*100,"% of the variance")

#Project x_train (scaled) and x_test (scaled) into 5 PC dimensions
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

#load LinearRegression into reg and then fit PCA data
reg2 = LinearRegression()
reg2.fit(X_train,y_train)
```

### 5.2 Evaluate the quantitative performance of the new model in terms of RMSE and R2 on the test set (10 points).

```python
predTrain = reg2.predict(X_train)
predTest = reg2.predict(X_test)
RMSETrain = np.sqrt(mean_squared_error(y_train,predTrain))
RMSETest = np.sqrt(mean_squared_error(y_test,predTest))
R2Train = r2_score(y_train,predTrain)
R2Test = r2_score(y_test,predTest)
print("RMSE_train = "+ str(RMSETrain))
print("RMSE_test = "+ str(RMSETest))
print("\n")
print("R2_train = "+ str(R2Train))
print("R2_test = "+ str(R2Test))
```
